```python
# Cell 1 -  Libraries import
import os 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import pickle
```

```python
import pandas as pd
df = pd.read_csv("cleaned_stock_prices.csv")
print("Shape:", df.shape)
df.head()
```

```python
# Cell 2 - Load CSV and basic inspect
path = 'cleaned_stock_prices.csv'
df = pd.read_csv(path)
print("Shape;", df.shape)
print("\nDtypes:")
print(df.dtypes)
print("\nFirst 6 rows:")
display(df.head(6))
```

```python
# Cell 3 -  Decide X (predictor) and y (target)
x_col = 'open'
y_col = 'close'
print("Pridictor (x):", x_col)
print("Target (y):", y_col)
print("\nSummary stats:")
display(df[[x_col, y_col]].describe())
```

```python
# Cell 4 - Preprocess: convert date, drop NaNs, show counts
df['date'] = pd.to_datetime(df['date'], errors='coerce')
before = len(df)
df_xy = df[[x_col, y_col, 'date', 'symbol']].dropna(subset=[x_col, y_col])
after = len(df_xy)
print(f"Rows before: {before}, after drooping NA in x/y: {after}, dropped: {before-after}")
display(df_xy.head(6))
```

```python
# Cell 5 - Train/test split (show counts)
x = df_xy[[x_col]].values
y = df_xy[y_col].values
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
print("Train rows:", len(x_train))
print("Test rows:", len(x_test))
```

```python
# Cell 6 - Fit Linear Regression and show model formula
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
coef = model.coef_[0]
intercept = model.intercept_
print(f"Fitted model: y = {coef:.6f} * x + {intercept:.6f}")
```

```python
# Cell 7 - Predict on test set and compute metrics
y_pred = model.predict(x_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("R-squared (test):", r2)
print("MSE (test):", mse)
```

```python
# Cell 8 - Show Actual vs Predicted first 12 rows
sample_df = pd.DataFrame({
    'open': x_test.flatten(),
    'actual_close': y_test,
    'predicted_close': y_pred,
    'residual': y_test - y_pred
})
display(sample_df.head(12))
```

```python
# Cell 9 - Plot scatter (test) + fitted line (show + save)
x_line = np.linspace(x.min(), x.max(), 200).reshape(-1,1)
y_line = model.predict(x_line)

plt.figure(figsize=(9,5))
plt.scatter(x_test, y_test)
plt.plot(x_line, y_line)
plt.xlabel('open')
plt.ylabel('close')
plt.title('Scatter (test) and Linear Fit')
plt.tight_layout()
plt.savefig('EDA_Analysis_2/regression_scatter.png')
plt.show()
print(r"Saved plot: C:\Users\Love\Desktop\EDA analysis\EDA_Analysis_2")
```

```python
# Cell 10 -  Residuals plot (show + save)
plt.figure(figsize=(9,5))
plt.scatter(y_pred, y_test - y_pred)
plt.axhline(0, linestyle='--')
plt.xlabel('Predicted close')
plt.ylabel('Residual (actul - predicted)')
plt.title('Residule vs Predicted')
plt.tight_layout()
plt.savefig('EDA_Analysis_2/regression_scatter.png')
plt.show()
print(r"Saved plot: C:\Users\Love\Desktop\EDA analysis\EDA_Analysis_2")
```

```python
# Cell 11 - Save model and prediction CSv
import pickle
with open('EDA_Analysis_2/linear_model_simple.pkl', 'wb') as f:
    pickle.dump(model, f)
print(r'Saved model to C:\Users\Love\Desktop\EDA analysis\EDA_Analysis_2')

out_df = sample_df.copy()
out_df.to_csv(
    r"C:\Users\Love\Desktop\EDA analysis\EDA_Analysis_2\predictions.csv", 
    index=False
)

print(r"Saved predictions to C:\Users\Love\Desktop\EDA analysis\EDA_Analysis_2\predictions.csv")
```
